## 2. Profiling Customer Churn dan Stay

```python
## grouping class ke churn dan stay cutomer type
class_group = ["Churn" if x>= 1 else "Stay" for x in cust_db['class']]
cust_db['class_group'] = class_group
```

```python
churn_cb = cust_db[cust_db['class_group']=="Churn"]
stay_cb = cust_db[cust_db['class_group']=="Stay"]
```

```python
## Average dari customer stay untuk total usage
column_select = ['total_day_charge','total_day_minutes', 'total_day_calls']

stay_cb[column_select].mean().round(2)
```

```python
## Average dari customer churn untuk total usage
churn_cb[column_select].mean().round(2)
```

```python
## Top kontributor state dari customer stay

stay_cb['state'].mode().round(2)
```

```python
## Top kontributor state dari customer churn
churn_cb['state'].mode().round(2)
```

```python
## korelasi total day charge untuk churn customer
abs(churn_cb[numeric_variable].corr()['total_day_charge']).sort_values(ascending=False).head(14)
```

```python
## korelasi total day charge untuk stay customer
abs(stay_cb[numeric_variable].corr()['total_day_charge']).sort_values(ascending=False).head(14)
```
